```python
def chromosome_collections(df, y_positions, height,  cent_buffer=0.3, ideo=False, **kwargs):
    """
    Yields BrokenBarHCollection of features that can be added to an Axes
    object.
    Parameters
    ----------
    df : pandas.DataFrame
        Must at least have columns ['chrom', 'start', 'end', 'color']. If no
        column 'width', it will be calculated from start/end.
    y_positions : dict
        Keys are chromosomes, values are y-value at which to anchor the
        BrokenBarHCollection
    ideo:default(False)- is this an idiogram plot? in which case, you can choose to shrink the height 
    of the centromeres
    
    cent_buffer:float amount to move up the centromere y_pos, with proportional decrease in
    height to give centromere more distinctive appearance
    
    height : float
        Height of each BrokenBarHCollection
    Additional kwargs are passed to BrokenBarHCollection
    """
    del_width = False
    
    if 'width' not in df.columns:
        del_width = True
        df['width'] = df['end'] - df['start']
        
        
    if ideo==True:
    
        for chrom, group in df.groupby('chrom'):
            
            for stain, g in group.groupby('gieStain'):
#                 print stain
                yrange = ''
                xranges =''
                if stain == 'acen':
                    yrange = (y_positions[chrom] + cent_buffer, height - (cent_buffer*2))
                    xranges = g[['start', 'width']].values
                else:
                    yrange = (y_positions[chrom], height)
                    xranges = g[['start', 'width']].values
                yield BrokenBarHCollection(
                    xranges, yrange, facecolors=g['colors'], **kwargs)
    
    if ideo==False:
        
        for chrom, group in df.groupby('chrom'):
            yrange = (y_positions[chrom], height)
            xranges = group[['start', 'width']].values
            yield BrokenBarHCollection(xranges, yrange, facecolors=group['colors'], **kwargs)
    
    
    if del_width:
        del df['width']
```

```python

       
def pos_2_df(x, chr_suff = True):
    
    """ make a dataframe from a ucsc coord or
    a coordinate from GS
    """
    try:
        chrom = x.split(':')[0].replace('chr','')
        start = int(x.split(':')[1].split('-')[0])
        end = int(x.split(':')[1].split('-')[1])
        name = "POS_{}_{}_{}".format(chrom, start, end)
       
    except:
        try:
            chrom = x.split('_')[1]
            start = int(x.split('_')[2])
            end = int(x.split('_')[3])
            name = "CNV_{}_{}_{}".format(chrom, start, end)

        except:
            print 'enter correct format for position'
    
    if chr_suff:
        data = ['chr' + chrom, start,end,name]
        cols = ['chrom', 'start', 'end', 'name']
        d_dict = dict(zip(cols,data))
        df = pd.DataFrame.from_records([d_dict])
        df = df[cols]
        return df, 'chr' + chrom
    if not chr_suff:
        data = [chrom.replace('chr',''), start,end,name]
        cols = ['chrom', 'start', 'end', 'name']
        d_dict = dict(zip(cols,data))
        df = pd.DataFrame.from_records([d_dict])
        df = df[cols]
        return df, chrom
        
```

```python
def calls_between_coords(x, start, end):
    
    """ 
    lambda function to subset calls between coordinates
    given that columns are named Start and End for 
    Chromosomal positions
    
    """
    if x.Start > start and x.Start < end:
        return True
    elif x.End < end and x.End > start:
        return True
    
    else:
        return False
```

```python
def get_hist_cov(Pos,  UUIDs, bt=False, pad=0, genome='/frazer01/publicdata/gatk_bundle_2.8/b37/human_g1k_v37_decoy_Sendai.genome', norm=False):
    
    """
    bt = False, if true expect pybedtools BedTool Object as input for read depth prediction
    cnv_ID = False  if true expects a cnv ID ex(CNV_CHR_START_END)
    by default expects pos in format of UCSC Browser = chr:start-end
    
    POS position in format chromosome, Start, End (default chr:start-end)
    
    
    UUID = iterable/list object of the samples desired
    
    pad = Desired amount of padding on either side of the position default=0
    
    genome= Desired Genome for use in padding the positions
    
    norm: if True, normalize read depth information to library size (under_development)
    
    """
    try:
        
        if bt == True:
            bt = Pos
        else:
            bt = pos_2_df(Pos, chr_suff=False)[0]
            bt = pybedtools.BedTool.from_dataframe(bt)

    except:
        print 'provide proper position and format flag'
        
    

    if pad<>0:
        bt = bt.slop(b=pad,g=genome)
        
    df_out = pd.read_table(bt.fn, names=['Chr', 'Start', 'End','ID'])
    for ids in UUIDs:
        
        bw_root = '/frazer01/projects/CARDIPS/pipeline/WGS/alignment_bwa/'
        bw_fn = bw_root+ ids + '/' + ids + '.mdup.bw'
        Command = ['/frazer01/home/djakubosky/bin/bwtool', 'extract', 'bed', bt.fn, bw_fn,'/dev/stdout']
        cov = subprocess.Popen(Command, stdout=subprocess.PIPE)
        df = pd.read_table(cov.stdout, names=['Chr', 'Start', 'End','ID','Features', 'hist'])
        df_out[ids] = df['hist'].values
    return df_out
```

```python
position = 'chr3:1230945-2340293'

def plot_chrom_pos_2(stuff):
#     try:
#         plt.cla()
    fig = plt.figure(figsize=(10,0.5))
    ax = fig.add_subplot(111)
    
    
    sns.set_style('ticks')
    # Height of each ideogram
    chrom_height = 1
    
    # Spacing between consecutive ideograms
    chrom_spacing = 1
    
    # Height of the region track. Should be smaller than `chrom_spacing` in order to
    # fit correctly
    region_height = 1.5
    
    # Padding between the top of a region track and its corresponding ideogram
    region_padding = 0.0
    
    # Decide which chromosomes to use
    figsize=(10,0.5)
    
    desired = pos_2_df(stuff)[1]
    
    print desired
    
    chromosome_list = ['chr%s' % i for i in range(1, 23) + ['X', 'Y']]
    
    chromosome_list=[i for i in chromosome_list if i==desired]
    
    # Keep track of the y positions for ideograms and regions for each chromosome,
    # and the center of each ideogram (which is where we'll put the ytick labels)
    ybase = 10
    chrom_ybase = {}
    region_ybase = {}
    chrom_centers = {}
    
    # Iterate in reverse so that items in the beginning of `chromosome_list` will
    # appear at the top of the plot
    for chrom in chromosome_list[::-1]:
        chrom_ybase[chrom] = ybase
        chrom_centers[chrom] = ybase + chrom_height / 2.
        region_ybase[chrom] = ybase - 0.5 - region_padding
        ybase += chrom_height + chrom_spacing
    
    # Read in ideogram.txt, downloaded from UCSC Table Browser
    ideo = pd.read_table(
        idiogram_loc,
        skiprows=1,
        names=['chrom', 'start', 'end', 'name', 'gieStain']
    )
    
    # Filter out chromosomes not in our list
    ideo = ideo[ideo.chrom.apply(lambda x: x in chromosome_list)]
    
    # Add a new column for width
    ideo['width'] = ideo.end - ideo.start
    
    # Colors for different chromosome stains
    color_lookup = {
        'gneg': (1., 1., 1.),
        'gpos25': (.6, .6, .6),
        'gpos50': (.4, .4, .4),
        'gpos75': (.2, .2, .2),
        'gpos100': (0., 0., 0.),
        'acen': (.8, .4, .4),
        'gvar': (.8, .8, .8),
        'stalk': (.9, .9, .9),
    }
    
    # Add a new column for colors
    ideo['colors'] = ideo['gieStain'].apply(lambda x: color_lookup[x])
    
    
    # Same thing for regions
    regions = pos_2_df(stuff)[0]
    regions = regions[regions.chrom.apply(lambda x: x in chromosome_list)]
    regions['width'] = regions.end - regions.start
    regions['colors'] = '#eb6841'
    
    
#     fig = plt.figure(figsize=figsize)
#     ax = fig.add_subplot(111)
    
    # Now all we have to do is call our function for the ideogram data...
    
    for collection in chromosome_collections(ideo, chrom_ybase, chrom_height, ideo=True, linewidths=2):
        ax.add_collection(collection)
    
    # Add the Marker for the location
    
    for collection in chromosome_collections(regions, region_ybase, region_height, alpha=1, linewidths=3, edgecolors=regions['colors']):
        ax.add_collection(collection)
    
    # Axes tweaking
    ax.set_yticks([chrom_centers[i] for i in chromosome_list])
    ax.set_yticklabels(chromosome_list, fontsize=12)
    
    # labels= [int(100*float(x.get_text()))for x in ax.get_xticklabels()[:-1]]
    # print ax.get_xlim()
    # l = [x.get_text() for x in ax.get_xticklabels()]
    # print l
    
    chrom_xticks = np.arange(0, chrom_dict[desired]+10000000, 50000000)
    chrom_xtick_labels = [str(int(x)) + ' MB' for x in chrom_xticks/10e5]
    
    ax.set_xticks(chrom_xticks)
    ax.set_xticklabels(chrom_xtick_labels)
    
    
    
    ax.axis('tight')
    # ax.set_ylim(ax.get_ylim()[0], ax.get_ylim()[1])
    ax.set_xlim(ax.get_xlim()[0]-1000000, ax.get_xlim()[1]+1000000)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('white')
    ax.spines['top'].set_visible(False)
    
    
    ax.get_xaxis().tick_bottom()
    for tic in ax.yaxis.get_major_ticks():
        tic.tick1On = tic.tick2On = False
```

```python
def plot_hist_2(site, genotypes, info, UUID=0, pad = 0, CNV_plot = False, regions = False):
    
    
    
    draw_chrom = plot_chrom_pos_2(site)
    
    
    
    """df is the hist cov df from get_hist_cov
    site is an ID of a specific Genome Strip Call
    uf,
    
    info: pandas df with minimum of ['Chr', 'Start', 'End'] columns: 
    a df of regions you would like plotted above the coverage graph
    
    pad= integer of padding on each side of site that you would like default=0
    
    CNV_plot, default: False,  if True plot regions as horizontal lines above the 
    coverage plot
    
    """
    
    df = get_hist_cov(site, UUID, pad=pad)
    subplot_num = 1
    df = df[df.ID==site]
    ## Get CNV_Coords
    
    if type(UUID)==list:
        subplot_num = len(UUID)
    
    fig, axes = plt.subplots(subplot_num, 1, figsize = (10, 5*subplot_num));
    
    start = int(site.split('_')[2])
    end = int(site.split('_')[3])
    
    # Make sure this has the right format
    Chr = 'chr' + site.split('_')[1]
    
    # Plot each of the histograms for each UUID
    
    
    try:
        
        padded_start = df.Start.tolist()[0]
        padded_end = df.End.tolist()[0]
    except: 
        
        padded_start = df.Start.values.tolist()[0]
        padded_end = df.End.values.tolist()[0]
        
    site_location = '{0}:{1}-{2}'.format(Chr, df.Start.tolist()[0], df.End.tolist()[0])
    Central_CNV = site
    
    ultimate_max_y = []
    for i,ids in enumerate(UUID):
        y= df[ids].values[0]
        y = y.split(',')
        y = np.where(np.array(y)=='NA',0.0,y)
        y = [float(x) for x in y] 
        x = np.arange(df.Start,df.End, 1)
        axes[i].plot(x,y);
        axes[i].fill_between(x,y);
        axes[i].set_title(ids);
        max_y = np.max(np.array(y))
        ultimate_max_y.append(max_y)
        
        mean_y = np.mean(np.array(y))
        median_y = np.median(np.array(y))
        min_y = np.min(np.array(y))

        formatted_Vals = ["{0:.2f}".format(l) for l in [mean_y, median_y, min_y, max_y]]
        formatted_Vals = formatted_Vals + [site_location]

        axes[i].set_ylim(0, max_y+20)
        axes[i].set_xlim(padded_start,padded_end)
        axes[i].text(padded_end + 1000, max_y, "Mean: {0}\nMedian: {1} \nMin: {2}\nMax: {3}\n{4}".format(*formatted_Vals), 
                     style='italic', horizontalalignment='left',
                     verticalalignment='top', bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})
    
        
        
        
        if CNV_plot==True:
            chrom_CNVS = info[info['Chr']== Chr]    
            contained = chrom_CNVS[chrom_CNVS.apply(lambda b: calls_between_coords(b,df.Start.tolist()[0], df.End.tolist()[0]), axis=1)==True]

            # Jitter the height of the hlines
            height = [6,8] *100
            count = 0

            for s,e,f in zip(contained.Start, contained.End, contained.Coords):
                genotype = genotypes.loc[f][ids]


                axes[i].text(((e-s)/2)+ s,max_y + height[count]+1,genotype, horizontalalignment='center');
                if int(genotype) < 2:

                    axes[i].hlines(max_y + height[count], s,e, colors='r', alpha=0.8);

                elif int(genotype) > 2:
                    axes[i].hlines(max_y + height[count], s,e, colors='g', alpha=0.8);
                else:
                    axes[i].hlines(max_y + height[count], s,e, colors='black', alpha=0.8);
                count +=1
        
        axes[i].vlines(start, 0, max_y, linestyles='dashed')
        axes[i].vlines(end, 0, max_y, linestyles='dashed')

        
    
    plt.suptitle(Central_CNV, fontsize=20);
    
    for i,ids in enumerate(UUID):
        axes[i].set_ylim(0, (max(ultimate_max_y) + max(ultimate_max_y)* 0.2))
    
    
    axes[len(UUID)-1].set_xlabel('Coordinates')
    plt.show()
    
    if CNV_plot==True:
        
        display(contained);
```
